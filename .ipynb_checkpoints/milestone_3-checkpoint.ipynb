```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
```

```python
df = pd.read_csv('Datafiniti_Amazon_Consumer_Reviews_of_Amazon_Products_May19.csv')
df.head(5)
```

```python
#check for rows with nulls

check_df = df[['reviews.rating' , 'reviews.text' , 'reviews.title' , 'reviews.username', 'reviews.doRecommend', 'reviews.didPurchase', 'reviews.numHelpful']]
print(check_df.isnull().sum()) 
```

```python
#Check for similarities
df.nunique()
```

```python
#del unused columns
del df["reviews.didPurchase"]
del df["reviews.doRecommend"]
del df["reviews.id"]
del df["reviews.numHelpful"]
df.head(5)
```

```python
#descriptive stats
df.describe()
```

```python
#delete rows with null
df = df.dropna(how='any')
df.count()
```

```python
#check most common username
perperson_rating=df["reviews.username"].value_counts()
perperson_rating.head(10)
```

```python
#clean the data "by"
perperson_rating= df["reviews.username"].replace(to_replace =r'By', value = '', regex = True).value_counts()
perperson_rating.head(10)
```

```python
print ("number of ratings : " + str(sum(perperson_rating)))
print ("number of users : " + str(len(perperson_rating)))
```

```python
print("Users giving bulk ratings (more than 20) : " + str(sum(perperson_rating >20)))
bulk_rating = perperson_rating[perperson_rating>20]
bulk_rating1 = sum(bulk_rating)
print ("Bulk ratings : " + str(bulk_rating1))
print ("Percentage of bulk ratings (including By Amazon and By Kindle customer) : " + str(bulk_rating1*100/sum(perperson_rating)))
print ("Percentage of bulk users : " + str(sum(perperson_rating>20)*100/len(perperson_rating)))
perperson_rating.value_counts().plot(kind='pie',figsize=(10,10), title='Ratings Per User')
```

```python
ratings_df = df.groupby(['reviews.rating'])
ratings_df.count()
```

```python
ratings_df = df.groupby(['brand','reviews.rating'])
ratings_df.count()
```

```python
#clean data to merge amazonbasics
df["brand"].replace(to_replace =r'Amazonbasics', value = 'AmazonBasics', regex = True, inplace = True)
ratings_df = df.groupby(['brand','reviews.rating'])
ratings_df.count()
```

```python
category_rating_df = df.groupby(['categories', 'reviews.rating'])
category_rating_df.count().head(10)
```

```python
category_rating_df = df.groupby(['primaryCategories', 'reviews.rating']).count()
category_rating_df.reset_index(inplace=True)
category_rating_df.head(10)
```

```python
pivot_df=category_rating_df.pivot(index="primaryCategories",columns="reviews.rating",values="id")
pivot_df.head()
```

```python
pivot_df.fillna(0)
```

```python
# plot bar
pivot_df.plot.bar(stacked=True)
plt.title("Categories Rating")
plt.xlabel("primary categories")
plt.ylabel("Number of Reviews")

```

```python
# Create bins in which to place values based upon TED Talk views
bins = [0,2,4,5]
# Create labels for these bins
group_names = ["Dectractors", "Passive", "promoters"]
```

```python
df["Rating_NPS"] = pd.cut(df["reviews.rating"],bins,labels=group_names)
df.head()
```

```python
group_by_ratings = df.groupby('Rating_NPS').size()
sorted_ratings = group_by_ratings.sort_values(ascending=False)
ratings_df = pd.DataFrame(sorted_ratings.rename('No of Users')).reset_index()

ratings_df
```

```python
#plot bar
number_of_users = ratings_df["No of Users"]
ratings_labels = ratings_df["Rating_NPS"]

plt.bar(ratings_labels, number_of_users, color='b', align="center")

# Give the chart a title, x label, and y label
plt.title("Promoter Score")
plt.xlabel("Type")
plt.ylabel("Number of Reviews")

plt.savefig("number_of_ratings.png")
plt.show()
```

```python
df['year'] = pd.DatetimeIndex(df['dateAdded']).year
df['month'] = pd.DatetimeIndex(df['dateAdded']).month
df.head(1)
```

```python
#explore time data
group_by_time = df.groupby('year').size()
sorted_df = group_by_time.sort_values(ascending=False)
time_df = pd.DataFrame(sorted_df.rename('No of Users')).reset_index()

time_df
```

```python
#try word cloud for text analysis

comment_words = '' 
stopwords = set(STOPWORDS) 
  
# iterate through the csv file 
for val in df["reviews.text"]: 
      
    # typecaste each val to string 
    val = str(val) 
  
    # split the value 
    tokens = val.split() 
      
    # Converts each token into lowercase 
    for i in range(len(tokens)): 
        tokens[i] = tokens[i].lower() 
      
    comment_words += " ".join(tokens)+" "
  
wordcloud = WordCloud(width = 800, height = 800, 
                background_color ='white', 
                stopwords = stopwords, 
                min_font_size = 10).generate(comment_words) 
  
# plot the WordCloud image                        
plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
  
plt.show() 
```

```python
#reduce word frequency

comment_words = '' 
stopwords = set(STOPWORDS) 
  
# iterate through the csv file 
for val in df["reviews.text"]: 
      
    # typecaste each val to string 
    val = str(val) 
  
    # split the value 
    tokens = val.split() 
      
    # Converts each token into lowercase 
    for i in range(len(tokens)): 
        tokens[i] = tokens[i].lower() 
      
    comment_words += " ".join(tokens)+" "
  
wordcloud = WordCloud(width = 600, height = 600, 
                background_color ='white', 
                stopwords = stopwords, 
                min_font_size = 20).generate(comment_words) 
  
# plot the WordCloud image                        
plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
  
plt.show() 
```
