```python
# Import the requisite packages
# Pandas for data manipulation
import pandas as pd
import numpy as np

# Matplotlib for basic plotting
import matplotlib.pyplot as plt

#import data
review_df = pd.read_csv("Datafiniti_Amazon_Consumer_Reviews_of_Amazon_Products_May19.csv")
review_df.head()
```

```python
#determine the count of users for each rating
group_by_ratings = review_df.groupby('reviews.rating').size()
sorted_ratings = group_by_ratings.sort_values(ascending=False)
ratings_df = pd.DataFrame(sorted_ratings.rename('No of Users')).reset_index()

ratings_df
```

```python
#create a barchart for count of users for each rating
number_of_users = ratings_df["No of Users"]
ratings_labels = ratings_df["reviews.rating"]

plt.bar(ratings_labels, number_of_users, color='b', align="center")

# Give the chart a title, x label, and y label
plt.title("Number of Ratings")
plt.xlabel("Rating")
plt.ylabel("Number of Reviews")

plt.savefig("number_of_ratings.png")
plt.show()
```

```python
#count the number of reviews by category
group_by_categories = review_df.groupby('primaryCategories')

category_count = group_by_categories.count().reset_index()
category_count
```

```python
#find average reviews of each category

category_rating_avg = group_by_categories.mean().reset_index()
category_rating_avg
```

```python
#create barchart of average reviews by category

average_rating = category_rating_avg["reviews.rating"]
categories = category_rating_avg["primaryCategories"]
x_axis = np.arange(len(categories))

plt.bar(x_axis, average_rating, color='b', align="center")
plt.xticks(x_axis, x_axis)

# Give the chart a title, x label, and y label
plt.title("Average Rating of Categories")
plt.xlabel("Category")
plt.ylabel("Average Rating")

# plt.legend(loc='upper right')

plt.savefig("avg_ratings_by_category.png")
plt.show()
```
